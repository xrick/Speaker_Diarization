```python
from scipy.cluster.vq import kmeans, vq, kmeans2, whiten
from numpy import array, reshape, zeros
from python_speech_features import mfcc, delta
from scipy.spatial import distance
from scipy.spatial.distance import pdist
import scipy.io.wavfile as wav
import wave
import matplotlib.pyplot as plt
import numpy as np
import math
import IPython.display as ipd
SIZE = 120
```

```python
ipd.Audio('.wav')
ipd.Audio('01.wav')
```

```python
(rate,sig) = wav.read("01.wav")
mfcc_feat = mfcc(sig,rate)

code_books = []
```

```python
whitened = whiten(mfcc_feat)
codebook, labeled_obs = kmeans2(data=whitened, k=SIZE)
code_books.append(codebook)
(rate,sig) = wav.read("02.wav")
mfcc_feat = mfcc(sig,rate)
whitened = whiten(mfcc_feat)
codebook, labeled_obs = kmeans2(data=whitened, k=SIZE)
code_books.append(codebook)
(rate,sig) = wav.read("03.wav")
mfcc_feat = mfcc(sig,rate)
whitened = whiten(mfcc_feat)
codebook, labeled_obs = kmeans2(data=whitened, k=SIZE)
code_books.append(codebook)
(rate,sig) = wav.read("03.wav")
mfcc_feat = mfcc(sig,rate)
whitened = whiten(mfcc_feat)
codebook, labeled_obs = kmeans2(data=whitened, k=SIZE)
code_books.append(codebook)
```

```python
(rate,sig) = wav.read(".wav")
test_feature_vector = mfcc(sig,rate)
num = 4
test_frame_num = test_feature_vector.shape[0]
print(test_feature_vector)
print(test_frame_num)
result = np.zeros(test_frame_num)
temp_min = np.zeros(num)
temp = 1
for i in range(1,test_frame_num):
    for j in range(0,num):
        for k in range(1,SIZE):
            x = (test_feature_vector[i,:], code_books[j][k, :])
            dist = sum(pdist(x, metric='euclidean'))
            if k == 1:
                temp_min[j] = dist
            else:
                if dist < temp_min[j]:
                    temp_min[j] = dist
    index = np.min(temp_min);
    result[i] = index
print(result)
smooth_size = 100
overlap = smooth_size//3
for i in range(test_frame_num//smooth_size -1):
    count = np.zeros(4)
    for j in range(1,smooth_size):
        if i == 0:
            count[int(result[int(i*smooth_size + j)])] = count[int(result[int(i*smooth_size + j)])] + 1
        else:
            count[int(result[int(i*smooth_size + j - overlap)])] = count[int(result[int(i*smooth_size + j - overlap)])] + 1
    index = count.max()
    for j in range(1,smooth_size):
        if i == 0:
            result[int(i*smooth_size + j)] = index;
        else:
            result[int(i*smooth_size + j - overlap)] = index;
plt.plot(result)
plt.show()
```
